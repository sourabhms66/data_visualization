```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
plt.title('comperison vk vs rs')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.plot(batsman['index'],batsman['V Kohli'],color = 'blue')
plt.title('comperison vk vs rs')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['V Kohli'],color = 'blue',linestyle='dotted')
plt.title('comperison vk vs rs')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed',marker='.')
plt.plot(batsman['index'],batsman['V Kohli'],color = 'blue',linestyle='dotted',marker='.')
plt.title('comperison vk vs rs')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed',marker='.',markersize=10)
plt.plot(batsman['index'],batsman['V Kohli'],color = 'blue',linestyle='dotted',marker='.')
plt.title('comperison vk vs rs')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed',marker='.',markersize=10,label='R S')
plt.plot(batsman['index'],batsman['V Kohli'],color = 'blue',linestyle='dotted',marker='.',label='V K')
plt.title('comperison vk vs rs')
plt.xlabel('season')
plt.ylabel('runs')
plt.legend()
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed',marker='.',markersize=10,label='R S')
plt.plot(batsman['index'],batsman['V Kohli'],color = 'blue',linestyle='dotted',marker='.',label='V K')
plt.title('comperison vk vs rs')
plt.xlabel('season')
plt.ylabel('runs')
plt.legend(loc='lower left')
```
