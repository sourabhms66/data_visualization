### Line plot
To plot line plot we'll import a dataset from pltly called gapminder, it's a dataset of copuntries population, gdp, and life expectancy in every 5 years

```python
gap = px.data.gapminder()
gap
```

Now we we wanna find out the life expantancy of indians from the year 1952 to letest

```python
gap[gap['country']=='India']
```

```python
ind = gap[gap['country']=='India']
```

```python
sns.scatterplot(data=ind,x=ind['year'],y=ind['lifeExp'])
```

Now to draw a line plot we can call lineplot() function from the seaborn library, which is a axes level function

```python
sns.lineplot(data=ind,x=ind['year'],y=ind['lifeExp'])
```

Drawing the same thing using figure level function, relplot

```python
sns.relplot(data=ind,x=ind['year'],y=ind['lifeExp'],kind='line')
```

Now if we wanna do the same thing for India ,Pakistan and china, we can do this

```python
gap[gap['country'].isin(['India','Pakistan','China'])]
```

```python
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
```

Now if we wanna plot a relplot()

```python
sns.relplot(data=temp_df,x=temp_df['year'],y=temp_df['lifeExp'],kind='line',hue='country') # figure level object
```

Now if we plot the same thing using line plot, we'll get the plot like the below, as we know that the relplot() is a figure level object and it has control all over the figure it can display the legend outside the plot. But lineplot() is an axes level object it doesn't have control all over the plot, only upto the axes so it displayed the legend inside the plot.

```python
sns.lineplot(data=temp_df,x=temp_df['year'],y=temp_df['lifeExp'],hue='country') # axes level object
```

Similarly we can do the same thing for different continent countries as well

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Australia'])]
```

```python
temp_df
```

```python
sns.relplot(data=temp_df,x=temp_df['year'],y=temp_df['lifeExp'],kind='line',hue='country')
```

We can further add one more parameter style to make distinguish between countries

```python
sns.relplot(data=temp_df,x=temp_df['year'],y=temp_df['lifeExp'],kind='line',hue='country',style='continent')
```

We can also add one more parameter called size it represents the thickness of the line

```python
sns.relplot(data=temp_df,x=temp_df['year'],y=temp_df['lifeExp'],kind='line',hue='country',style='continent',size='continent')
```

```python
sns.displot(data=tips,x=tips['tip'],kind='kde',hue='sex')
```

### Rugplot

Plot marginal distributions by drawing ticks along the x and y axes.

This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(data=tips,x=tips['total_bill'])
```

So we need to use a axis level function here because we are trying plot two plots in one object

```python
sns.kdeplot(data=tips,x=tips['total_bill'])
sns.rugplot(data=tips,x=tips['total_bill'])
```

### 2. Matrix Plot

- Heatmap
- Clustermap

Here we're trying to plot a heatmap(plot the data in retangular form depending on the density of the data points) of life expectancy each year of all the countries

```python
gap
```

So to pplot this we have convert the data to wide form from long form, so we can get all the values of each year, and we do that using pivot function.

```python
gap.pivot(index='country', columns='year',values='lifeExp')
```

Now our job is to represent this whole dataframe in the form of heatmap grid, to do that let's first store this dataframe

```python
temp_df = gap.pivot(index='country', columns='year',values='lifeExp')
```

heatmap is axes level function. and it doesn't have a corresponding figure level function. And now we just need to pass the dataframe in this function.

```python
sns.heatmap(temp_df)
```

Since this is a axes level function, now if we want we can use plt to change the size of the plot

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

Now we do further data analysis, like if we wanna find out the life expectancy of asia, we would first create a dataframe for asia then do the same thing on it.

```python
gap[gap['continent'] == 'Asia']
```

```python
gap[gap['continent'] == 'Asia'].pivot(index='country',columns='year',values='lifeExp')
```

```python
temp_df = gap[gap['continent'] == 'Asia'].pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

We can further adjust the parameters like, if we wanna show the numbers like

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True)
```

and further if we wanna show a little gap between the values we can use another parameter like linewidth

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidths=0.5)
```

If we wanna change the color map we can use the cmap parameter

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidths=0.5,cmap='viridis')
```

### how to make changes in figure level function

```python
sns.displot(data=tips,x=tips['tip'],kind='kde',hue='sex',fill=True)
```

Now if we wanna change using matplotlib, it won't work because they are different figures

```python
plt.figure(figsize=(15,10))
sns.displot(data=tips,x=tips['tip'],kind='kde',hue='sex',fill=True)
```

As we can see no changes has been made. we can use a parameter called hight, assiging a value will get the job done, it'll increase the width internally automatically

```python
sns.displot(data=tips,x=tips['tip'],kind='kde',hue='sex',fill=True,height=10,aspect=2)
```
