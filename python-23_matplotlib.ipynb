For more information about linestyle look at the cell below or visit here https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
import matplotlib.pyplot as plt
import numpy as np

linestyle_str = [
     ('solid', 'solid'),      # Same as (0, ()) or '-'
     ('dotted', 'dotted'),    # Same as (0, (1, 1)) or ':'
     ('dashed', 'dashed'),    # Same as '--'
     ('dashdot', 'dashdot')]  # Same as '-.'

linestyle_tuple = [
     ('loosely dotted',        (0, (1, 10))),
     ('dotted',                (0, (1, 1))),
     ('densely dotted',        (0, (1, 1))),
     ('long dash with offset', (5, (10, 3))),
     ('loosely dashed',        (0, (5, 10))),
     ('dashed',                (0, (5, 5))),
     ('densely dashed',        (0, (5, 1))),

     ('loosely dashdotted',    (0, (3, 10, 1, 10))),
     ('dashdotted',            (0, (3, 5, 1, 5))),
     ('densely dashdotted',    (0, (3, 1, 1, 1))),

     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))]


def plot_linestyles(ax, linestyles, title):
    X, Y = np.linspace(0, 100, 10), np.zeros(10)
    yticklabels = []

    for i, (name, linestyle) in enumerate(linestyles):
        ax.plot(X, Y+i, linestyle=linestyle, linewidth=1.5, color='black')
        yticklabels.append(name)

    ax.set_title(title)
    ax.set(ylim=(-0.5, len(linestyles)-0.5),
           yticks=np.arange(len(linestyles)),
           yticklabels=yticklabels)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.spines[:].set_visible(False)

    # For each line style, add a text annotation with a small offset from
    # the reference point (0 in Axes coords, y tick value in Data coords).
    for i, (name, linestyle) in enumerate(linestyles):
        ax.annotate(repr(linestyle),
                    xy=(0.0, i), xycoords=ax.get_yaxis_transform(),
                    xytext=(-6, -12), textcoords='offset points',
                    color="blue", fontsize=8, ha="right", family="monospace")


fig, (ax0, ax1) = plt.subplots(2, 1, figsize=(10, 8), height_ratios=[1, 3])

plot_linestyles(ax0, linestyle_str[::-1], title='Named linestyles')
plot_linestyles(ax1, linestyle_tuple[::-1], title='Parametrized linestyles')

plt.tight_layout()
plt.show()
```

# Stacked bar chart
A stacked bar chart is a type of bar chart that represents categorical data with rectangular bars stacked on top of each other. Each bar in the chart represents a category, and the length of the bar corresponds to the value of that category. However, instead of showing only one value per category as in a regular bar chart, a stacked bar chart shows multiple values for each category, where each value is represented by a segment of the bar.



This means one bar represent the total runs of a batsman of all the season different color represent runs in one season

```python
plt.bar(df['batsman'],df['2017'])
plt.bar(df['batsman'],df['2016'],bottom=df['2017'])
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'])
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],label='2015')
plt.legend()
plt.show()
```

# Histogram

A histogram is a graphical representation of the distribution of numerical data. It consists of a series of adjacent rectangles, or bars, where the width of each bar represents a range of values, and the height of each bar represents the frequency or count of data points falling within that range. Histograms are commonly used to visualize the distribution of continuous data, such as measurements or observations.

Key features of a histogram include:

- Bins or intervals: The range of values is divided into discrete intervals, called bins or buckets. Each bin represents a range of values, and data points falling within that range are counted.

- Frequency or count: The height of each bar represents the frequency or count of data points falling within the corresponding bin. Higher bars indicate a higher frequency of data points in that range.

- Continuous data: Histograms are suitable for visualizing continuous data, where the values can take any numeric value within a certain range.

Histograms are useful for understanding the distributional characteristics of a dataset, such as the central tendency, spread, skewness, and presence of outliers. They provide insights into the shape of the data distribution, including whether it is symmetric, skewed to the left or right, bimodal, or multimodal.

Histograms are widely used in various fields, including statistics, data analysis, and data visualization, to explore and analyze the underlying patterns and structures within datasets. They are also commonly used in quality control, finance, biology, and many other domains for understanding and interpreting data distributions.
